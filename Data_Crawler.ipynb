```python
import requests
from itertools import product
import pandas as pd
import time
```

```python
def get_product_data(text,pages):
    try:
        import requests
        from bs4 import BeautifulSoup
        Data = {}
        for page in range(1, pages+1):
            url = f"https://www.flipkart.com/search?q={text}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={page}"
            req = requests.get(url)
            req.status_code
            soup = BeautifulSoup(req.text, "html.parser")
            D = []
            for i in soup.findAll("div",{"class":"_1xHGtK _373qXS"}):
                d={}
                for j in i.findAll("a",{"class":"_2UzuFa"}):
                    d["href"] = "https://www.flipkart.com"+j["href"]
                    # D.append(j["href"])
                for j in i.findAll("img",{"class":"_2r_T1I"}):
                    d["img"] = j["src"]
                for j in i.findAll("div",{"class":"_2WkVRV"}):
                    d["company"] = j.text
                for j in i.findAll("a",{"class":"IRpwTa"}):
                    d["title"] = j["title"]
                for j in i.findAll("div",{"class":"_30jeq3"}):
                    d["price"] = j.text
                for j in i.findAll("div",{"class":"_3I9_wc"}):
                    d["ogprice"] = j.text
                for j in i.findAll("div",{"class":"_3Ay6Sb"}):
                    d["discount"] = j.text
                D.append(d)
                Data[f"page_data_{page}"] = D
        return(Data)
    except Exception as e:
        raise e
```
